```python
num_lockers = 100
lockers = [True]*num_lockers
```

```python

for student in range(1,num_lockers,1):
    #print(student)
    for student_touch in range(student,num_lockers,student):
        print(student_touch)
        if lockers[student_touch-1] == True:
            lockers[student_touch-1] == False
        else:
            lockers[student_touch-1] == True
#print(sum(lockers))
print(lockers)
```
