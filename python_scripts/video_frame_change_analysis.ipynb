```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def get_files():
    path = "video_auto_analysis"
    dir_list = os.listdir(path)
    dir_list = [x[:-4] for x in dir_list]
    return dir_list


def avg_shot_duration(df):
    return df['Length (frames)'].mean() / 30
def avg_shot_duration_before_10s(df):
    return df[df['End Time (seconds)']<11]['Length (frames)'].mean() / 30
def avg_shot_duration_before_30s(df):
    return df[df['End Time (seconds)']<31]['Length (frames)'].mean() / 30
def avg_shot_duration_before_60s(df):
    return df[df['End Time (seconds)']<61]['Length (frames)'].mean() / 30
def avg_shot_duration_before_120s(df):
    return df[df['End Time (seconds)']<121]['Length (frames)'].mean() / 30
def avg_shot_duration_before_300s(df):
    return df[df['End Time (seconds)']<300]['Length (frames)'].mean() / 30
def avg_shot_duration_before_600s(df):
    return df[df['End Time (seconds)']<600]['Length (frames)'].mean() / 30
def longest_shot(df):
    return df['Length (frames)'].max() / 30
def shortest_shot(df):
    return df['Length (frames)'].min() / 30

def comp_all_stats(id,name,df,columns):
    stats_df = pd.DataFrame([[
        id
        , name
        , avg_shot_duration(df)
        , avg_shot_duration_before_10s(df)
        , avg_shot_duration_before_30s(df)
        , avg_shot_duration_before_60s(df)
        , avg_shot_duration_before_120s(df)
        , avg_shot_duration_before_300s(df)
        , avg_shot_duration_before_600s(df)
        , longest_shot(df)
        , shortest_shot(df)
        ]],
        columns=columns
    )
    return stats_df
    

# def main():
files = get_files()
main_df = pd.DataFrame(
    columns=[
        'video_id'
        ,'Video Name'
        ,'Avg Shot Duration'
        ,'Avg Shot Duration (<10s)'
        ,'Avg Shot Duration (<30s)'
        ,'Avg Shot Duration (<1m)'
        ,'Avg Shot Duration (<2m)'
        ,'Avg Shot Duration (<5m)'
        ,'Avg Shot Duration (<10m)'
        ,'Longest Shot'
        ,'Shortest Shot'
        ]
)
for file in enumerate(files):
    # each id is 
    curr_id = file[0]+1
    curr_name = file[1]
    curr_df_stats = []
    curr_df = pd.read_csv(f"./video_auto_analysis/{curr_name}.csv",skiprows=[0])
    curr_df_stats = comp_all_stats(curr_id,curr_name,curr_df,main_df.columns)
    main_df = pd.concat([main_df,curr_df_stats])


# if __name__ == "__main__":
#     main()
```
